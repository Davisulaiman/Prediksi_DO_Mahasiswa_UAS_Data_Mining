```python
def determine_gender_from_name(name):
    """
    Menentukan gender berdasarkan nama (heuristic)
    """
    # Nama-nama yang umum untuk perempuan
    female_indicators = [
        'putri', 'sari', 'dewi', 'nur', 'esa', 'sinta', 'ulfa', 'anissa', 'zahrah',
        'neli', 'citra', 'shanniyah', 'carli', 'aisyah', 'saniyyah', 'dian', 'fanni',
        'amirah', 'tiesya', 'raden', 'charissa', 'rana', 'keisya', 'atika', 'sophina',
        'imelda', 'natasya', 'evelyn', 'hikmah', 'federika', 'resyaliana', 'deva',
        'aurel', 'kenia', 'hesi', 'linia', 'lidya', 'azzahra', 'diosi', 'intan',
        'sallaa', 'rheby', 'hana', 'herlita', 'nadya', 'adelia', 'tis', 'filya',
        'anis', 'fidia', 'aulia', 'primanda', 'khayla', 'najwa', 'merischa', 'rahma',
        'waridhania', 'dinda', 'khaylilla', 'khalisa', 'anggun', 'cynthia', 'qonita',
        'ludvie', 'azka', 'nurmelizah', 'shintia', 'prycillia', 'naifah', 'tasya',
        'tamara', 'nurul', 'leksa', 'atikah', 'aisyah', 'nayla', 'anjelia', 'lovien',
        'widya', 'mulan', 'angelita', 'revanny', 'nania', 'afifah', 'dwi', 'lia',
        'ayuni', 'naila', 'adiva'
    ]

    name_lower = name.lower()
    for indicator in female_indicators:
        if indicator in name_lower:
            return 'P'

    return 'L'  # Default ke laki-laki jika tidak ada indikator perempuan
```

```python
def generate_realistic_features(npm, name, cohort_year):
    """
    Generate fitur-fitur realistis untuk setiap mahasiswa
    """

    # Tentukan semester saat ini berdasarkan angkatan
    if cohort_year == 2022:
        current_semester = random.choice([6, 7])  # Senior
    elif cohort_year == 2023:
        current_semester = random.choice([4, 5])  # Junior
    else:  # 2024
        current_semester = random.choice([2, 3])  # Sophomore

    # Data demografis
    gender = determine_gender_from_name(name)
    age = 18 + (2025 - cohort_year) + random.randint(-1, 2)

    # Faktor sosioekonomis
    family_income_options = ['Rendah', 'Menengah', 'Tinggi']
    family_income_weights = [0.35, 0.50, 0.15]  # Distribusi realistis
    family_income = random.choices(family_income_options, weights=family_income_weights)[0]

    # Beasiswa (lebih tinggi untuk income rendah)
    if family_income == 'Rendah':
        scholarship_prob = 0.6
    elif family_income == 'Menengah':
        scholarship_prob = 0.25
    else:
        scholarship_prob = 0.1

    scholarship = 1 if random.random() < scholarship_prob else 0

    # Status pekerjaan
    work_status_options = ['Tidak Bekerja', 'Part-time', 'Full-time']
    work_status_weights = [0.65, 0.30, 0.05]
    work_status = random.choices(work_status_options, weights=work_status_weights)[0]

    # Asal daerah
    origin_options = ['Dalam Kota', 'Luar Kota', 'Luar Provinsi']
    origin_weights = [0.45, 0.35, 0.20]
    origin = random.choices(origin_options, weights=origin_weights)[0]

    # Latar belakang pendidikan
    high_school_options = ['SMA Negeri', 'SMA Swasta', 'SMK']
    high_school_weights = [0.55, 0.30, 0.15]
    high_school_type = random.choices(high_school_options, weights=high_school_weights)[0]

    # Skor masuk (UTBK/tes masuk)
    entry_score = max(60, min(95, np.random.normal(75, 10)))

    entry_score = round(entry_score, 1)

    # Hitung risk score
    risk_score = 0
    if family_income == 'Rendah': risk_score += 2
    if work_status == 'Full-time': risk_score += 3
    if scholarship == 0 and family_income == 'Rendah': risk_score += 2
    if entry_score < 70: risk_score += 2
    if high_school_type == 'SMK': risk_score += 1
    if origin == 'Luar Provinsi': risk_score += 1
    if age > 22: risk_score += 1

    high_risk = risk_score >= 4

    # Performance akademik
    base_gpa = entry_score / 25  # Konversi ke skala 4.0

    # Adjustment berdasarkan faktor risiko
    if family_income == 'Rendah': base_gpa -= random.uniform(0, 0.25)
    if work_status == 'Full-time': base_gpa -= random.uniform(0.15, 0.4)
    if origin == 'Luar Provinsi': base_gpa -= random.uniform(0, 0.15)

    # Generate IPK dengan variasi per semester
    semester_gpas = []
    current_gpa = max(1.0, min(4.0, base_gpa + random.uniform(-0.25, 0.25)))

    for sem in range(current_semester):
        # Variasi natural per semester
        variation = random.uniform(-0.15, 0.15)

        # Trend berdasarkan risk factors
        if high_risk:
            trend = random.uniform(-0.08, 0.03)  # Cenderung menurun
        else:
            trend = random.uniform(-0.03, 0.08)  # Cenderung stabil/naik

        current_gpa = max(0.5, min(4.0, current_gpa + trend + variation))
        semester_gpas.append(round(current_gpa, 2))

    avg_gpa = round(np.mean(semester_gpas) if semester_gpas else 0, 2)
    gpa_trend = round(semester_gpas[-1] - semester_gpas[0], 2) if len(semester_gpas) > 1 else 0

    # Kehadiran
    if high_risk:
        avg_attendance = random.randint(55, 80)
        attendance_consistency = random.uniform(8, 20)
    else:
        avg_attendance = random.randint(78, 98)
        attendance_consistency = random.uniform(2, 12)

    # Aktivitas LMS
    if high_risk:
        total_lms_hours = random.randint(20, 100) * current_semester
    else:
        total_lms_hours = random.randint(80, 200) * current_semester

    avg_lms_per_sem = round(total_lms_hours / current_semester, 1) if current_semester > 0 else 0

    # Mata kuliah diulang
    if high_risk:
        total_retakes = random.randint(1, 5)
    else:
        total_retakes = random.randint(0, 2)

    # Status saat ini
    if high_risk and random.random() < 0.12:
        current_status = 'Dropout'
        dropout_semester = random.randint(2, current_semester)
    elif avg_gpa < 2.0:
        current_status = 'Peringatan Akademik'
        dropout_semester = None
    else:
        current_status = 'Aktif'
        dropout_semester = None

    # Risk category
    if risk_score >= 4:
        risk_category = 'Tinggi'
    elif risk_score >= 2:
        risk_category = 'Sedang'
    else:
        risk_category = 'Rendah'

    # Early warning flag
    early_warning_flag = 0
    if len(semester_gpas) >= 2:
        if np.mean(semester_gpas[:2]) < 2.5 or avg_attendance < 70 or total_retakes >= 2:
            early_warning_flag = 1
    elif len(semester_gpas) == 1:
        if semester_gpas[0] < 2.5 or avg_attendance < 70:
            early_warning_flag = 1

    # Intervention recommendation
    recommendations = []
    if avg_gpa < 2.5:
        recommendations.append('Bimbingan Akademik')
    if family_income == 'Rendah' and scholarship == 0:
        recommendations.append('Bantuan Finansial')
    if avg_attendance < 75:
        recommendations.append('Konseling Motivasi')
    if work_status == 'Full-time':
        recommendations.append('Penyesuaian Jadwal')
    if not recommendations:
        recommendations.append('Monitoring Rutin')

    intervention_recommendation = ', '.join(recommendations)

    return {
        'npm': npm,
        'nama': name,
        'angkatan': cohort_year,
        'semester_saat_ini': current_semester,
        'jenis_kelamin': gender,
        'usia': age,
        'penghasilan_keluarga': family_income,
        'beasiswa': scholarship,
        'status_pekerjaan': work_status,
        'asal_daerah': origin,
        'jenis_sma': high_school_type,
        'skor_masuk': entry_score,
        'ipk_rata_rata': avg_gpa,
        'trend_ipk': gpa_trend,
        'kehadiran_rata_rata': avg_attendance,
        'konsistensi_kehadiran': round(attendance_consistency, 1),
        'total_jam_lms': total_lms_hours,
        'rata_jam_lms_per_semester': avg_lms_per_sem,
        'total_mengulang': total_retakes,
        'skor_risiko': risk_score,
        'kategori_risiko': risk_category,
        'status_saat_ini': current_status,
        'semester_dropout': dropout_semester,
        'flag_peringatan_dini': early_warning_flag,
        'rekomendasi_intervensi': intervention_recommendation
    }
```

```python
def create_comprehensive_dataset():
    """
    Membuat dataset lengkap untuk prediksi dropout
    """
    print("🚀 Memulai pembuatan dataset prediksi dropout mahasiswa...")

    # Ekstrak data mahasiswa
    students_2022, students_2023, students_2024 = extract_student_data()

    all_students = []

    print("📊 Memproses data angkatan 2022...")
    # Proses angkatan 2022
    for student_data in students_2022:
        # Extract NPM dan nama
        parts = student_data.split(' ', 1)
        npm = parts[0]
        name = parts[1] if len(parts) > 1 else "Unknown"

        student_features = generate_realistic_features(npm, name, 2022)
        all_students.append(student_features)

    print("📊 Memproses data angkatan 2023...")
    # Proses angkatan 2023
    for student_data in students_2023:
        parts = student_data.split(' ', 1)
        npm = parts[0]
        name = parts[1] if len(parts) > 1 else "Unknown"

        student_features = generate_realistic_features(npm, name, 2023)
        all_students.append(student_features)

    print("📊 Memproses data angkatan 2024...")
    # Proses angkatan 2024
    for student_data in students_2024:
        parts = student_data.split(' ', 1)
        npm = parts[0]
        name = parts[1] if len(parts) > 1 else "Unknown"

        student_features = generate_realistic_features(npm, name, 2024)
        all_students.append(student_features)

    # Konversi ke DataFrame
    df = pd.DataFrame(all_students)

    # Sorting berdasarkan NPM
    df = df.sort_values('npm').reset_index(drop=True)

    return df

def generate_summary_statistics(df):
    """
    Generate statistik ringkasan dataset
    """
    print("\n" + "="*60)
    print("📈 RINGKASAN STATISTIK DATASET")
    print("="*60)

    print(f"Total mahasiswa: {len(df)}")
    print(f"\nDistribusi per angkatan:")
    print(df['angkatan'].value_counts().sort_index())

    print(f"\nDistribusi status saat ini:")
    print(df['status_saat_ini'].value_counts())

    print(f"\nDistribusi kategori risiko:")
    print(df['kategori_risiko'].value_counts())

    print(f"\nDistribusi jenis kelamin:")
    print(df['jenis_kelamin'].value_counts())

    print(f"\nStatistik IPK:")
    print(f"Rata-rata IPK: {df['ipk_rata_rata'].mean():.2f}")
    print(f"Median IPK: {df['ipk_rata_rata'].median():.2f}")
    print(f"IPK terendah: {df['ipk_rata_rata'].min():.2f}")
    print(f"IPK tertinggi: {df['ipk_rata_rata'].max():.2f}")

    print(f"\nStatistik kehadiran:")
    print(f"Rata-rata kehadiran: {df['kehadiran_rata_rata'].mean():.1f}%")
    print(f"Kehadiran terendah: {df['kehadiran_rata_rata'].min():.1f}%")
    print(f"Kehadiran tertinggi: {df['kehadiran_rata_rata'].max():.1f}%")

    print(f"\nMahasiswa dengan peringatan dini: {df['flag_peringatan_dini'].sum()} ({df['flag_peringatan_dini'].mean()*100:.1f}%)")

    print(f"\nMahasiswa dropout: {len(df[df['status_saat_ini'] == 'Dropout'])} ({len(df[df['status_saat_ini'] == 'Dropout'])/len(df)*100:.1f}%)")
```

```python
# ====================================================================
# MODIFIED MAIN EXECUTION - SAVE TO GOOGLE DRIVE FOLDER
# ====================================================================

import os

def save_to_drive_folder(dataset, folder_name='uas_data_mining', filename='dataset_prediksi_dropout_TI_2022_2024.csv'):
    """
    Simpan dataset ke Google Drive dalam folder tertentu
    """
    try:
        # Deteksi environment
        is_colab = 'google.colab' in str(get_ipython()) if 'get_ipython' in globals() else False

        if is_colab:
            # Path untuk Google Colab
            drive_folder = f'/content/drive/MyDrive/{UAS_DATA_MINING}'

            # Cek apakah Google Drive sudah di-mount
            if not os.path.exists('/content/drive'):
                print("📱 Mounting Google Drive...")
                from google.colab import drive
                drive.mount('/content/drive')

        else:
            # Path untuk environment lokal
            drive_folder = folder_name

        # Buat folder jika belum ada
        os.makedirs(drive_folder, exist_ok=True)

        # Path file lengkap
        file_path = os.path.join(drive_folder, filename)

        # Simpan dataset
        dataset.to_csv(file_path, index=False, encoding='utf-8')

        # Info hasil
        file_size = os.path.getsize(file_path) / 1024  # KB

        print(f"\n✅ Dataset berhasil disimpan!")
        print(f"📁 Lokasi: {file_path}")
        print(f"📊 File berisi {len(dataset)} records dengan {len(dataset.columns)} kolom")
        print(f"📏 Ukuran file: {file_size:.1f} KB")

        return file_path

    except Exception as e:
        print(f"❌ Error menyimpan ke drive: {str(e)}")

        # Fallback: simpan ke current directory
        fallback_filename = filename
        dataset.to_csv(fallback_filename, index=False, encoding='utf-8')
        print(f"💾 Fallback: File tersimpan sebagai '{fallback_filename}' di current directory")

        return fallback_filename
```
